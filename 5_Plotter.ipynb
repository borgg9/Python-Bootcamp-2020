```python
def createdata():
    """Que precios quieres descargar de quandl"""
    # si no está aqui repetir: https://s3.amazonaws.com/quandl-production-static/end_of_day_us_stocks/ticker_list.csv
    lista_stock = []
    stock = 1
    while stock != 'fin':
        stock = input('Que acciones quieres analizar?: ')
        lista_stock.append("EOD/" + stock.upper())
    lista_stock = lista_stock[:(len(lista_stock)-1)] 
    return lista_stock
```

```python
# Crea DataFrame con los precios de las acciones que piden
acciones = quandl.get(createdata(), column_index='1')
acciones
```

```python
def plotter(plot):
    if(plot == 1):
        finalplot = acciones.iplot(kind='scatter')
    elif(plot == 2):
        finalplot = acciones.iplot(kind='scatter', mode='markers', symbol='x', colorscale='paired')
    elif(plot == 3):
        finalplot = acciones.iplot(kind='bar')
    elif(plot == 4):
        finalplot = acciones.iplot(kind='hist')
    elif(plot == 5):
        finalplot = acciones.iplot(kind='box')
    elif(plot == 6):
        finalplot = acciones.iplot(kind='surface')
    else:
        finalplot = print('Selecciona un gráfico entre 1 y 6')
    return finalplot
```

```python
def plotter2(plot):  
    col = input('Cuantas columnas quieres imprimir? (1 ó 2): ')
    col = int(col)
    
    if (col == 1):
        col = input("Que columna quieres imprimir? (EOD/XXX - Open): ")
        if (plot == 1):
            finalplot = acciones[col].iplot(kind='scatter')
        elif (plot == 2):
            finalplot = acciones[col].iplot(kind='scatter', mode='markers', symbol='x', colorscale='paired')
        elif (plot == 3):
            finalplot = acciones[col].iplot(kind='bar')
        elif (plot == 4):
            finalplot = acciones[col].iplot(kind='hist')
        elif (plot == 5):
            finalplot = acciones[col].iplot(kind='box')
        elif (plot == 6) or (plot == 7):
            finalplot = print('bubble y surface necesitan más de un elemento')
        else:
            finalplot = print('Selecciona un gráfico entre 1 y 7')
    
    if (col == 2):
        print('Que 2 acciones quieres ver?: ')
        x = input("Acción 1? (EOD/XXX - Open): ")
        y = input("Acción 2? (EOD/XXX - Open): ")
        if (plot == 1):
            finalplot = acciones[[x, y]].iplot(kind='scatter')
        elif (plot == 2):
            finalplot = acciones[[x, y]].iplot(kind='scatter', mode='markers', symbol='x', colorscale='paired')
        elif (plot == 3):
            finalplot = acciones[[x, y]].iplot(kind='bar')
        elif (plot == 4):
            finalplot = acciones[[x, y]].iplot(kind='hist')
        elif (plot == 5):
            finalplot = acciones[[x, y]].iplot(kind='box')
        elif (plot == 6):
            finalplot = acciones[[x, y]].iplot(kind='surface')
        elif (plot == 7):
            size = input('Tamaño para bubble plot')
            finalplot = acciones.iplot(kind='bubble', x=x, y=y, size=size)
        else:
            finalplot = print('Selecciona un gráfico entre 1 y 7')  
    
    else:
        finalplot = print('Selecciona un gráfico entre 1 ó 2')
    return finalplot
```

```python
def main(cat):
    if(cat == 1):
        print("Selecciona el tipo de gráfico (1-6)\n1. Line\n2. Scatter\n3. Bar\n4. Histogram\n5. Box\n6. Surface")
        plot = int(input())
        output = plotter(plot)
    elif(cat == 2):
        print("Selecciona el tipo de gráfico (1-7)\n1. Line\n2. Scatter\n3. Bar\n4. Histogram\n5. Box\n6. Surface\n7. Bubble")
        plot = int(input())
        output = plotter2(plot)
    else:
        print('Por favor introduce 1 o 2')
```
